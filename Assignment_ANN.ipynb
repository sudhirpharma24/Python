```python
# Data Split up - Training and Testing
```

```python
x = df.drop(columns=['letter'])
y = df['letter']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
# Model Building
# Sequence of Model in following order
#1.Initialize model
#2.Add hidden layer
#3.Add output layer
#4.Establish the connection between the layers
#5.Fit the data
```

```python

ann = Sequential()

ann.add(Dense(units = 10, activation = 'relu'))
ann.add(Dense(units = 1, activation = 'sigmoid'))

ann.compile(optimizer = 'adadelta', loss = 'binary_crossentropy', metrics =['accuracy'])

ann.fit(xtrain,ytrain,epochs = 100)
```

```python
ypred = ann.predict(xtest)
ypred = ypred > 0.5
ypred
```

```python
print(classification_report(ytest, ypred))
# Initial Model gives accuracy of 4%
```

```python
ann.evaluate(xtrain,ytrain)
```

```python
# Hyper Parameter Tuning
```

```python
x.shape
```

```python
def hyper(hp):
    model = Sequential()

    # Loop to add layers
    for i in range(hp.Int('num_layers', 10, 25)):
        units = hp.Int(f'units_{i}', 15, 200, 20)  # Number of units(neuron) in the layer
        activation = hp.Choice(f'activation_{i}', ['relu', 'tanh', 'sigmoid'])  # Activation function

        if i == 0:
            # First layer needs input dimension
            model.add(Dense(units, activation=activation, input_dim=16))
        else:
            # Other layers
            model.add(Dense(units, activation=activation))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(
        optimizer=hp.Choice('optimizer', ['adam', 'rmsprop', 'sgd']),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
tuner = kf.RandomSearch(hyper,
                        objective = 'val_accuracy',
                        max_trials = 3)
```

```python
tuner.search(xtrain,ytrain,epochs = 5, validation_data = (xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
# Best parameter for first model
```

```python
tuned_model = tuner.get_best_models(num_models=1)[0]
# Final model buidling with first best model
```

```python
tuned_model.fit(xtrain,ytrain,epochs=25, validation_data=(xtest,ytest))
```

```python
tuned_model.evaluate(xtrain,ytrain)
# Accuracy of train dataset
```

```python
tuned_model.evaluate(xtest,ytest)
# Accuracy of Test dataset
```
